```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn import datasets
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mice_data = pd.merge(mouse_metadata,study_results)
# Display the data table for preview
mice_data
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

# Mean of the tumor volume
mice_mean = mice_data_cleaned.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]

# Median of the tumor volume
mice_med = mice_data_cleaned.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]

# variance of the tumor volume
mice_var = mice_data_cleaned.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]

# Standard Deviation of the tumor volume
mice_std = mice_data_cleaned.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]

# SEM of the tumor volume
mice_sem = mice_data_cleaned.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]


# Assemble the resulting series into a single summary dataframe.
mice_summ = pd.DataFrame({
    "Mean": mice_mean,
    "Median": mice_med,
    "Variance": mice_var,
    "Standard Deviation": mice_std,
    "SEM": mice_sem
})

mice_summ
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

mice_summ2=mice_data_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby(["Drug Regimen"]).agg(['mean','median','var','std','sem'])
mice_summ2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
mice_vol = []



# Start by getting the last (greatest) timepoint for each mouse
l_m = mice_data_cleaned.groupby(["Mouse ID"])["Timepoint"].max()
l_m_df = pd.DataFrame(l_m)


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
lm2 = pd.merge(mice_data_cleaned, l_m_df, on = ("Mouse ID", "Timepoint"), how = "right")
lm2
```

```python
# Put treatments into a list for a for loop (and later for plot labels)

    
# Determine outliers using upper and lower bounds
   

# Create empty list to fill with tumor vol data (for plotting)
#Run a loop and append tumor values by regimen
for x in regimen:
    regi_df = lm2.loc[lm2["Drug Regimen"] == x,"Tumor Volume (mm3)"]
    mice_vol.append(regi_df)
    
# Calculate the IQR and quantitatively determine if there are any potential outliers.
m_quart = regi_df.quantile([0.25,0.5,0.75])

m_l_q = m_quart[0.25]

m_u_q = m_quart[0.75]

m_iqr = m_u_q - m_l_q

m_ubound = m_u_q + (1.5*m_iqr)

m_lbound = m_l_q - (1.5*m_iqr)

# Determine outliers using upper and lower bounds   
outliers = regi_df.loc[(regi_df < m_lbound)|(regi_df > m_ubound)]

# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, bp1 = plt.subplots()
bp1.set_title('Tumor Volumes at Last Timepoints (with outliers)')
bp1.set_ylabel('Tumor Volume (mm3)')
bp1.boxplot(mice_vol,regimen)
plt.show()
   
print(outliers)  
```

## Correlation and Regression

```python
# Calculate the correlation coefficient 
# for mouse weight and average tumor volume for the Capomulin regimen
m_corr = st.pearsonr(capmean_vol['Weight (g)'], capmean_vol['Tumor Volume (mm3)'])

m_corr
print(f"Correlation is {round(m_corr[0],2)}")

#linear regression model 
x_values = capmean_vol['Weight (g)']
y_values = capmean_vol['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)

plt.scatter(capmean_vol['Weight (g)'],capmean_vol['Tumor Volume (mm3)'], color='orange')
plt.title('Average tumor volume vs. Mouse weight')
plt.xlabel(' Mouse weight(g)')
plt.ylabel('Average tumor volume(mm3)')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq, (20,38), fontsize = 12, color = "red")
```
